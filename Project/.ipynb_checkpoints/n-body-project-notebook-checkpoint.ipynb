# Phys 257 Project: n-body simulation (using fft)

N-body simulations are widely used in studying different types of phenomenon and in this project we try to develop a code to do n-body simulation in 3-dimension space. 
(Note: I've included the key results and outputs here but for running the simulation it would be better to run the python code)

The method used here utilises the power of FFT to reduce the computation requirements and significantly decrease computation time. 

```python
# -*- coding: utf-8 -*-
"""
Created on Sun Dec  10 23:41:44 2020

@author: hussa
"""

import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from IPython import display
from IPython.display import Image
plt.ion()
```

First step in achieving this is to generate gravitational potential from a single particle in 3d space. This is done through the **get_V** function below.

```python
def get_V(n, softening = 3.0, self_shield = 2.0): # Similar procedure as assignment 7 but in 3d and with softening
	V=np.zeros((n,n,n)) # Forming the grid
	x=np.arange(n)
	x[n//2:]=x[n//2:]-n
	xx,yy,zz=np.meshgrid(x,x,x) # Defining x, y values on the grid
	xx[0,0,0]  = 1.0 
	yy[0,0,0]  = 1.0            # Taking care of the origin by re-setting it
	zz[0,0,0]  = 1.0 

	V = -1/np.sqrt(xx*xx+yy*yy+zz*zz) # Potential in 3d 

	r = np.linspace(-self_shield, self_shield, int(self_shield)*2+1) # Softening isnt supercritical here because self shielding takes care of the close range interaction
	for i in r:
		for j in r:
			for k in r:
				if np.sqrt(i**2+j**2+k**2)<self_shield:         # Self shielding
					V[int(i),int(j),int(k)] = -(1/self_shield)  # Flattening the potential around particle location to make self-force=0
					continue
				if np.sqrt(i**2+j**2+k**2)<softening: 		# Softening
					V[int(i),int(j),int(k)] = -(3.*softening**2-(i**2+j**2+k**2))/(2.*softening**3) # Potential inside solid sphere
	return V
```

The way the function works is as follows:

1) Generate a grid and acquire positions in it.

2) Calculate 1/r where r is distance from the origin. Locating center at origin helps when we will later convolve this with several particles.

3) In order to prevent the influence of a particular potential on itself, we then flatten the potential near the origin to make the force/accelaration there zero (softening is also implemented to soften the changes at smaller scale).

The potential looks like follows:

```python
z = get_V(100,self_shield=20) # I purposely used large self shielding, ideally a couple grid points is enough
plt.imshow(z[:,:,50]); plt.colorbar()
```

Since the particle is located at the origin we see that the potential at the origin is wrapped around and appears at the corners.

Next up, we define a class named Particle for defining the particle properties, initial conditions and evolution with time.

The **__ init__** function defines the particles based on the set initial conditions

The **get_forces** function is for acquiring the potential at a given configuration of particles and convolve them with corresponding postions to obtain the potential in the entire space. It also allows to set periodic boundaries which enables the particles on the edges to wrap around to the other side.

And finally, the **update** function is for updating the particle positions for a given time step. It uses **leap frog** for accurately calculating the step size and direction based on velocity calculation at half-step.

```python
class Particle:
	def __init__(self, x, v, g_size, m=1):
		self.x = x 			 	 	 	# Initial positions
		self.v = v 			 	 	 	# Initial velocities
		self.m = m 			 	 	 	# Paticle masses
		self.n = x.shape[0] 	 		# No. of particles
		self.f = np.empty(x.shape)		# Forces
		self.nn = g_size 				# Size of the 3d grid
	def get_forces(self,x,periodic):
		mask = np.zeros((self.nn,self.nn,self.nn)) 
		mask[x[:,0],x[:,1],x[:,2]] = self.m     # Setting position of particles onto the grid 
		self.f=0                                # Initialize force
		if periodic:
			V = get_V(self.nn)
			V_fft = np.fft.rfftn(V)	
			V_new = np.fft.irfftn(np.fft.rfftn(mask)*V_fft)*(-1)
		else:
			V = get_V(self.nn*2)
			V_fft = np.fft.rfftn(V)	
			mask=np.pad(mask,(0,mask.shape[0]),mode = 'constant') # Zero Padding
			V_new = np.fft.irfftn(np.fft.rfftn(mask)*V_fft)*(-1)  # The extra (-1) is added early for the force calculation 
			V_new=V_new[:self.nn,:self.nn]		                  # Trimming the edges
		fx,fy,fz = np.gradient(V_new)                             # Calculating the force components  
		self.f = np.transpose([fx[x[:,0],x[:,1],x[:,2]],fy[x[:,0],x[:,1],x[:,2]],fz[x[:,0],x[:,1],x[:,2]]])
		return (V_new[x[:,0],x[:,1],x[:,2]]+V[0,0,0])*self.m  #The V[0,0,0] is to remove the potential at the position from itself
	def update(self,dt,periodic):
		# Implementing leapfrog        
		x_temp = self.x+0.5*self.v*dt       # Moving by half step
		xxx = x_temp.copy()
		xxx = np.asarray(xxx,dtype = 'int')
		if periodic:
			xxx = xxx%self.nn               # wrap particles going outside the grid around
		U = self.get_forces(xxx,periodic)   # Obtaining the force at half step location
		v_temp = self.v + 0.5*dt*self.f
		self.x = self.x + dt*v_temp         # Updating location
		if periodic:
			self.x = self.x%self.nn         # wrap particles going outside the grid around
		self.v = self.v + dt*self.f         # Updating velocity
		K = 0.5*self.m*np.sum(v_temp[:,0]**2+v_temp[:,1]**2+v_temp[:,2]**2) # Kinetic energy
		return K,np.sum(U)/2.               # 1/2. factor is to compensate for double counting of potential energy
```

## Part 1

The first part is for performing a check on our settings of self shielding. The particle with zero initial velocity should not move provided it doesnt see any force from its potential. This is confirmed for the results included below. 

An important aspect in setting the initial conditions is the calculation of the time step **dt** size. I do this by calculating the maximum by computing the maximum velocity attainable by the particle for a given value of self shielding.  

```python
# ---------------------------------------------------------------------------
# Part 1
# ---------------------------------------------------------------------------

itter = 300			# Number of iterations/steps
n = 1				# Number of particles
g_size = 64			# Grid size
self_shield = 2  	# Self shielding (default)      
dt = self_shield**(-3/2)*0.05 # Time step based on the maximum possible velocity given self shielding is used
K = np.zeros(itter)
U = np.zeros(itter)

x = g_size//2+np.zeros((n,3)) 	# Setting particle location to center of grid
v = np.random.rand(n,3)*0.0		# Setting initial velocity to zero

pp = Particle(x,v,g_size) # Generating object of class particle for given initial conditions
for j in range(itter):
	for i in range(5):		
		K[j],U[j] = pp.update(dt,False)	# Updating the particle position for fixed timestep and non-periodic B.C.	
	# Recording energies after every 5th step	
	plt.figure(figsize=(8,8))
	plt.title('Single Particle (iteration : '+str(j)+')')
	plt.plot(pp.x[:,0],pp.x[:,1],'o')   # Visualizing particle in 2d
	plt.axis([0,g_size,0,g_size])
	display.display(plt.gcf())
	display.clear_output(wait=True)
	plt.pause(0.5)
	
	# Visualizing in 3d
#	fig=plt.figure(figsize=(10,10))#Create 3D axes
#	try: ax=fig.add_subplot(111,projection="3d")
#	except : ax=Axes3D(fig) 
#	ax.set_xlim3d(0,g_size)
#	ax.set_ylim3d(0,g_size)
#	ax.set_zlim3d(0,g_size)	
#	ax.scatter(pp.x[:,0],pp.x[:,1], pp.x[:,2],color="royalblue",marker="o")
#	ax.set_xlabel("x-coordinate",fontsize=14)
#	ax.set_ylabel("y-coordinate",fontsize=14)
#	ax.set_zlabel("z-coordinate",fontsize=14)
#	ax.set_title("Single particle (starting from rest)\n",fontsize=20)
#	ax.legend(loc="upper left",fontsize=14)
#	plt.savefig('images/part1'+str(j)+'.png', dpi=200)	
#	plt.close()	
```

## Part 2

For the second part, we analyse the dynamics of the simplest case of 2 particles undergoing circular motions.

Basic physics behind this that for the correct initial conditions (position/velocity), two particles can undergo circular motion. For the case I am considering, I take two identical particles located 64/3 ~ 21.3 grid cell units apart.

For this system the potential energy = -1/21.3 ~ **-0.0468**

For circular motion we can easily derive that Kinetic energy = -0.5*Potential energy. Thus the expression for the initial velocity is given as follows (for G=m=1)

$$V_{init} = (1/(2d))$$
where d is the distance between the 2 particles

```python
# ---------------------------------------------------------------------------
# Part 2
# ---------------------------------------------------------------------------

itter = 200					# Number of steps
n = 2						# Number of particles
g_size = 64					# Grid size
self_shield = 2				# Not very important since orbiting particles wont come close to each other
#dt = softening**(-1.5)*0.05
dt = 0.5 					# For the separation I chose, dt=1 is small enough to avoid issues and speeds up the sim

E = np.zeros(itter)
K = np.zeros(itter)
U = np.zeros(itter)

x = [[g_size/3,g_size/2,g_size/2],[2*g_size/3,g_size/2,g_size/2]] # Particle position
x = np.asarray(x)
#v=np.random.rand(n,3)*0
v = [[0,1,0],[0,-1,0]]
v = np.asarray(v)

# For circular orbit, v = sqrt(Gm^2/(R(2m))) for identical masses, here i set G=m=1, hence v=sqrt(1/(2d))
alpha = np.sqrt(0.5/(x[1,0]-x[0,0]))		
v = v*alpha

pp = Particle(x,v,g_size)
for j in range(itter):
	for i in range(5):		
		K[j],U[j] = pp.update(dt,False)		
	E[j] = (K[j]-U[j])

	# Plotting in 2d for more clear visualization
	plt.figure(figsize=(8,8))
	plt.title('Two Particle circular orbit (iteration : '+str(j)+')')
	plt.plot(pp.x[:,0],pp.x[:,1],'o')
	plt.axis([0,g_size,0,g_size])
#	plt.savefig('images/part2'+str(j)+'.png', dpi=50)	
##	plt.close()
	display.display(plt.gcf())
	display.clear_output(wait=True)
	plt.pause(0.5)
```

Output of the above code can be seen to undergo perfect circular motion. Below is a gif with the assembled plots showing the two particles undergoing circular motion. 
#### Note: The track in the gif appears elliptical and not circular because of the shape of the image being rectangular. Its a perfect circle when resized to have the x,y axis of same size 

```python
Image(url='images/part2.gif') 
```

We can also track the energy of the particles and compare with expected results. The plots below clearly shows the expected trend for the energies, i.e.: **(PE = -2*\KE)** and **(TE = -KE)**

The plot for the potential energy shows **minor fluctuations**, this stems from minor changes in potential when switching grid cell boundaries. Upon averaging the value of potential we get value very close to **-0.0468** which is what we initially computed. Likewise, the value of kinetic energy falls close to **+0.0234**. 

#### This confirms that the leapfrog method conserves energy.  

```python
plt.title('Energy profile')
plt.plot(E,label='Total energy')
plt.plot(K,label='Kinetic energy')
plt.plot(-U,label='Potential energy')
plt.legend()
plt.show()
```

## Part 3 (periodic)
We can repeat this process but now with periodic boundary conditions which is more imprtant when simulating the universe. We create the periodic boundary by ignoring the padding step in **get_V** function. This causes the potential to wrap around which is what is intended for making it periodic.

```python
itter = 200
n = 100000
g_size = 128
self_shield = 2				
dt = self_shield**(-1.5)*0.05

x = np.random.rand(n,3)*g_size
v = np.random.rand(n,3)*0
K = np.zeros(itter)
U = np.zeros(itter)

pp = Particle(x,v,g_size)

for j in range(itter):
	for i in range(5):		
		K[j],U[j] = pp.update(dt,True)		# True: Setting periodic boundary
	
	#fig=plt.figure(figsize=(10,10))#Create 3D axes
	#try: ax=fig.add_subplot(111,projection="3d")
	#except : ax=Axes3D(fig) 
	#ax.set_xlim3d(0,g_size)
	#ax.set_ylim3d(0,g_size)
	#ax.set_zlim3d(0,g_size)	
	#ax.scatter(pp.x[:,0],pp.x[:,1], pp.x[:,2],color="royalblue",marker=".",s=.02)
	#ax.set_xlabel("x-coordinate",fontsize=14)
	#ax.set_ylabel("y-coordinate",fontsize=14)
	#ax.set_zlabel("z-coordinate",fontsize=14)
	#ax.set_title("3d n-body periodic (iteration: "+str(j)+" )\n",fontsize=20)
	#plt.savefig('images/part3(1)'+str(j)+'.png', dpi=50)	
	#plt.close()
```

## Part 4

Having observed all the previous results, we are now almost completely equipped to simulate universe formations. The only difference from the periodic case here is that the way we define particles and their locations is not so straight forward. 

For setting up scale invariance in 3-d, we need the mass fluctuation to follow k^-3. The procedure I follow is to first create a grid with all grid cell values sampled from a uniform distribution. The fourier space realisation part is a bit more subtle and involves computing the value of sqrt(k^-3) for every cell in inverse fourier space. Noise is added to this using the fourier transform of the white noise we initially got.

Then, the inverse fourier transform of this gives us the density map with the k^-3 fluctuation.

```python
# ----------------------------------------------------------------------------		
# Part 4
# ----------------------------------------------------------------------------		

g_size = 128
grid = np.random.rand(g_size,g_size,g_size) # Grid, White noise
grid_fft = np.fft.fftn(grid) 

x=np.fft.fftfreq(g_size)
kx,ky,kz=np.meshgrid(x,x,x)   # k definition
kx[0,0,0]=1.0
ky[0,0,0]=1.0
kz[0,0,0]=1.0 
Ak = (kx**2+ky**2+kz**2)**(-3/4) 	# Constructing the fourier realisation
Ak[0,0,0]=0.0						# Setting k = 0 to 0 (as opposed to 1/k^3 which tend to infinity)
Ak = Ak*grid_fft
m_den = np.real(np.fft.ifftn(Ak)) 	# Obtaining density in position/real space

plt.title('Density fluctuations for k^-3 case')
plt.imshow(m_den[:,:,64]) 			# Showing one slice

prob = (m_den-np.min(m_den))/(np.max(m_den)-np.min(m_den)) # Rescaling density to probablity

x = []
for i in range(g_size):
	for j in range(g_size):
		for k in range(g_size):
			if prob[i,j,k] > np.random.rand(): 	# Pickking points based on probablity
				x.append([i,j,k]) 
print(len(x))				
x = np.asarray(x)				
```

The density map can then be rescaled to give a probablity map which can be used to populate the grid with particles. These then set the initial position of our particles.

```python
g_size = 128
itter = 100
self_shield = 2				
dt = self_shield**(-1.5)*0.05
n = len(x)
v=np.random.rand(n,3)*0
v = np.asarray(v)

pp = Particle(x,v,g_size)

for j in range(itter):
	print(j)
	for i in range(5):		
		K,U = pp.update(dt,True)		
	
	#fig=plt.figure(figsize=(10,10))#Create 3D axes
	#try: ax=fig.add_subplot(111,projection="3d")
	#except : ax=Axes3D(fig) 
	#ax.set_xlim3d(0,g_size)
	#ax.set_ylim3d(0,g_size)
	#ax.set_zlim3d(0,g_size)	
	#ax.scatter(pp.x[:,0],pp.x[:,1], pp.x[:,2],color="royalblue",marker=".",s=.02)
	#ax.set_xlabel("x-coordinate",fontsize=14)
	#ax.set_ylabel("y-coordinate",fontsize=14)
	#ax.set_zlabel("z-coordinate",fontsize=14)
	#ax.set_title("Scale invariant power spectrum\n",fontsize=20)

	#plt.savefig('images/part4'+str(j)+'.png', dpi=100)	
	#plt.close()
```
