# Finding Gravitational waves

```python
import numpy as np
from matplotlib import pyplot as plt
import h5py
import glob
import os

# Code directly borrowed from the simple_reader
def read_template(filename):
    dataFile=h5py.File(filename,'r')
    template=dataFile['template']
    th=template[0]
    tl=template[1]
    return th,tl

def read_file(filename):
    dataFile=h5py.File(filename,'r')
    dqInfo = dataFile['quality']['simple']
    qmask=dqInfo['DQmask'][...]

    meta=dataFile['meta']
    #gpsStart=meta['GPSstart'].value
    gpsStart=meta['GPSstart'][()]
    #print meta.keys()
    #utc=meta['UTCstart'].value
    utc=meta['UTCstart'][()]
    #duration=meta['Duration'].value
    duration=meta['Duration'][()]
    #strain=dataFile['strain']['Strain'].value
    strain=dataFile['strain']['Strain'][()]
    dt=(1.0*duration)/len(strain)

    dataFile.close()
    return strain,dt,utc
```

Above is code borrowed from the simple_reader shared on github. Next I define some functions that will be use repeatedly in calculations.

First up is the **Window function**. I use the Hanning window used in class and add a central flat region. For the below plot I use 1000 points defined such that 0.75x1000 are spanned by the flat region 

```python
# Fucntion which is like 1-cos at edges but like box in middle
# Apparently it has a name :P Tukey window function
def window_func(n, t): 
	y = np.ones(n) # Defining a box extending the entire range
	m = int((1-t)*n/2)
	cos_edge = 0.5-0.5*np.cos(np.linspace(0,1,2*m)*2*np.pi)
	for i in range(m): # Smoothening out the transition at the edges by multiplying with cosine
		y[i] = y[i]*cos_edge[i]
		y[-(i+1)] = y[-(i+1)]*cos_edge[-(i+1)] 
	return y 

wind = window_func(1000,0.75) 

plt.figure(figsize=(12,6))      
plt.title('Window function-Tukey')
plt.ylabel('f(x)')
plt.grid(alpha=0.75)
plt.xlabel('x')    
plt.plot(wind)
plt.show()
plt.close()
```

Next up is **smoothener** for smoothinening the power-spectrum by averaging using neighbouring points. Argument **pix** defines the number of points to be used for averaging, including the current point. 

```python
# Function for smoothening of power-spectrum by averaging using neighboring points
def smoothener(x, pix):
	t = np.zeros(len(x))
	t[:(pix+1)//2]  = 1 	# Box car
	t[-(pix+1)//2:] = 1
	t = t/pix 			# Normalising
	return np.fft.ifft(np.fft.fft(x)*np.fft.fft(t),len(x))
```

Finally a **cross_corr** function to return cross correlation between x and y 

```python
# Function to calculate cross-corelation like the one in class
# Using the fourier method saves time because of fft being of time complexity N*log(N)
def cross_corr(x, y): 
	return np.fft.ifft(np.fft.fft(x)*np.conj(np.fft.fft(y)))
```

## Part A

To start off with development of noise model I start-off with preparing a window using the previously defined function. I set the span of the flat region to be **90%** of total. The choice of this particular value was made to avoid suppressing the signal by the window at the edges. Smaller fraction of flat region tends to affect the signal. (Effect of the window edge becomes more apprent when we look at the matched filter plot)

I then use this window on both the data from Hanford/Livingston as well as their templates before computing their fourier transforms (this helps reduce the spectral leakage).

```python
# Trying out windowing using a customized hanning window 
# which has a central flat region, like a box car with cosine edges
# 0.90 argument below is the fraction of the window occupied by the central flat part
wind = window_func(len(strain_h),0.90) 

plt.figure(figsize=(12,6))      
plt.title('Window function-Tukey')
plt.ylabel('f(x)')
plt.grid(alpha=0.75)
plt.xlabel('x')    
plt.plot(wind)
plt.show()
plt.close()

#wind = np.linspace(0,1,len(strain_h))*2*np.pi
#wind = 0.5*(1-np.cos(wind))

# Now strating off with windowing the data and template	
strain_h = strain_h*wind
strain_l = strain_l*wind
strain_h_fft1 = np.fft.fft(strain_h)
strain_l_fft1 = np.fft.fft(strain_l)

th = th*wind
tl = tl*wind
th_fft1 = np.fft.fft(th)
tl_fft1 = np.fft.fft(tl)
```

Next, I use the fourier transform of the windowed data as an estimate of the error present in the data. Using this I then define the noise matrix. Doing so I also make the assumption that noise is stationary enabling me to significantly simply the noise modelling procecss. Thus in fourier space I have a diagonal matrix for the noise matrix.

I then apply smoothening to this diagonal matrix. I do so by averaging the points with their neighbouring points to balance out any random drops (from the Gaussian noise) in the way power is distributed over the frequencies. Reason why I explicitly talk about 'drops' is because the effect of a small element is to weight its corresponding frequency higher than its supposed to and with smoothening we try and reduce that.

```python
# Now I define the noise matrix assuming the noise in data is stationary
# This implies that the noise matrix in fourier space is diagonal/uncorrelated
# Hence I take the fft of the strain as my sigma and define N as its square
N_h	= np.abs(strain_h_fft1)**2
N_l	= np.abs(strain_l_fft1)**2

# Smoothening the power spectrum by averaging using neighbouring 4 points, 2 on either side 	
sN_h = smoothener(N_h,25) # Number needs to be an odd integer
sN_l = smoothener(N_l,25) # To evenly smooth either side
```

In addition to smoothening, we can also compare the current noise matrix to one before smoothening to accomodate any peaks that get flattened out due to the averaging which otherwise are infact present.

```python
# Taking max of the original noise over smooth noise to recover peaks
smN_h = np.maximum(sN_h,N_h)		
smN_l = np.maximum(sN_l,N_l)
```

We can see the effect of these operations on our noise matrix below

```python
plt.figure(figsize=(12,6))      
plt.title('Noise model (Livingston)')
#	plt.ylabel('Strain')
plt.grid(alpha=0.75)
plt.xlabel('Index/time')    
plt.loglog(freq,abs(N_l),label='noise')
plt.loglog(freq,abs(smN_l),label='smooth (max) noise')
plt.loglog(freq,abs(sN_l),label='smooth noise')
plt.legend()	
plt.show()
plt.close()
```

Next I use the noise matrix to **pre-whiten** the  data. This is done to see/pick out the frequencies that matter most, which are the ones coming from the actual signal. This can be done in fourier space by diving with the square-root of the diagonal elements of the noise matrix.  

Here I use the smoothened noise matrix over the 'max' one because the results of matched filter seem to fare better with the prior, than when I use the later. One justification for this observation I can give is due to my choice of number of neighbours I average over. I use relatiely fewer points for the averaging which prevent from the peaks from getting averaged out while averaging out the Gaussian noise. 

```python
th_fft = th_fft1/np.sqrt(sN_h)
tl_fft = tl_fft1/np.sqrt(sN_l)
strain_h_fft = strain_h_fft1/np.sqrt(sN_h)
strain_l_fft = strain_l_fft1/np.sqrt(sN_l)
```

```python
plt.figure(figsize=(12,6))      
plt.title('Whitened template/data')
#	plt.ylabel('Strain')
plt.grid(alpha=0.75)
plt.xlabel('Index/time')    
plt.loglog(freq,abs(th_fft),label='whitened template (Hanford)')
plt.loglog(freq,abs(strain_h_fft),label='whitened datat (Hanford)')
plt.legend()	
plt.show()
plt.close()
```

Now that we have whitened the data and have our noise matrix ready we can proceed to performing the matched filter.

***
## Part B

```python
# 	PART B
# ---------------------------------------------------------------------------

mf_h_fft = np.conj(th_fft)*strain_h_fft
mf_l_fft = np.conj(tl_fft)*strain_l_fft
mf_h = np.fft.ifft(mf_h_fft)
mf_l = np.fft.ifft(mf_l_fft)

plt.figure(figsize=(12,6))      
plt.title('Matched filter')
plt.ylabel('Cross-correlation')
plt.grid(alpha=0.75)
plt.xlabel('Index/shift')    
#	plt.xscale('log')
plt.plot((mf_h)[:],label='Hanford')
plt.plot((mf_l)[:],label='Livington')
plt.legend()
plt.show();plt.close()
```

This the result of performing matched filter using above computed pre-whitened template and data. We can see clearly to clear peaks several sigma larger that the noise for data from both detectors. The pinch-off at the center is an artifact of the matched filter we used. The choice of large fraction of flat region helped in containing the pinch-off far from the signal which was kind of expected as the LIGO data is well centered. If the signal were near the edges this would then have been an issue.

***
## Part C

Signal to noise ratio (SNR) calculations: I use the ratio of the peak value from the matched filter to the standard deviation (avoiding the peak and the pinch-off regions). 

I do this for both the detector data and also report the combined SNR as well.


```python
# 	PART C 
# ---------------------------------------------------------------------------

# SNR from the results of matched filters of Hanford and livingston
#	print('argmax = ',np.argmax(abs(mf_h)),np.argmax(abs(mf_l)))
snr_l = np.abs(mf_l)/np.std(mf_l[np.argmax(abs(mf_l))+10:40000]) # Using a smaller chunck to calc. std to
snr_h = np.abs(mf_h)/np.std(mf_h[np.argmax(abs(mf_h))+10:40000]) # avoid the signal peak and window edge
snr = np.sqrt(np.max(snr_l)**2+np.max(snr_h)**2)
print('SNR Livingston: ', np.max(snr_l)) 
print('SNR Hanford '	, np.max(snr_h)) 
print('Combined event SNR:', snr)
```

This shows that with some basic use of knowledge about matched filters we can get the observation of signal to near 22sigma which is great. In the actual LIGO analysis they use a more involved averaging method called Welch averag rather than simply averaging like done here which is one of the things that could be implemented here for better results.

## Part D

The SNR can be derived from the analytic expression for matched filter as well. 
\begin{equation}
m = \frac{A^{T}N^{-1}d}{A^{T}N^{-1}A}
\end{equation}
For prewhitened elements, this is simply:
\begin{equation}
m = \frac{A^{T}d}{A^{T}A}
\end{equation}

```python
# 	PART D 
# ---------------------------------------------------------------------------

#SNR from analytic expression of whitened template and data
th_w = np.fft.ifft(th_fft) # Windowed template
tl_w = np.fft.ifft(tl_fft) # Windowed template
strain_h_w = np.fft.ifft(strain_h_fft) # WIndowed strain
strain_l_w = np.fft.ifft(strain_l_fft) # WIndowed strain

snr_la = np.abs(cross_corr(tl_w,strain_l_w)/np.dot(tl_w,tl_w)**0.5)*(len(strain_h))**0.5 # Additional factor of sqrt(N) in denominator
snr_ha = np.abs(cross_corr(th_w,strain_h_w)/np.dot(th_w,th_w)**0.5)*(len(strain_h))**0.5 # to take care of normalization
snra = np.sqrt(np.max(snr_la)**2+np.max(snr_ha)**2)

plt.figure(figsize=(12,6))      
plt.title('Matched filter (Ana)')
plt.ylabel('SNR')
plt.grid(alpha=0.75)
plt.xlabel('Index/shift')    
#	plt.xscale('log')
plt.plot(snr_ha,label='Hanford')
plt.plot(snr_la,label='Livington')
plt.legend()
plt.show();plt.close()

print('SNR Livingston: ', np.max((snr_la))) 
print('SNR Hanford '	, np.max((snr_ha))) 
print('Combined event SNR:',snra)
```

The results here match pretty well with the previous ones which had SNR ~ 22. Reason for the mismatch is the portion near the center. In the SNR calculation using scatter we avoided this which wasnt done in the analytic approach. A more pronounced effect can be observed by going back to the window function and choosing a narrow flat region. Doing so will increase the mismatch even further compared to what we see now.

## Part E

Calculating the frequency that marks the half way point for the weighting from the power spectrum. I do this for the noise model and the template model (before whitening)

```python
# 	PART E 
# ---------------------------------------------------------------------------

# COmputing 50% weight frequency
#psc_h = np.cumsum(np.abs(strain_h_fft1))/np.sum(np.abs(strain_h_fft1)) # PS cumulative livingston (fraction)
#psc_l = np.cumsum(np.abs(strain_l_fft1))/np.sum(np.abs(strain_l_fft1))
psc_l = np.cumsum(np.abs(sN_l))/np.sum(np.abs(sN_l))
psc_h = np.cumsum(np.abs(sN_h))/np.sum(np.abs(sN_h))

tpsc_h = np.cumsum(np.abs(th_fft1))/np.sum(np.abs(th_fft1)) # template PS cumulative livingston (fraction)
tpsc_l = np.cumsum(np.abs(tl_fft1))/np.sum(np.abs(tl_fft1))

fh = freq[np.argwhere(psc_h>=0.5)[0][0]]
fl = freq[np.argwhere(psc_l>=0.5)[0][0]]
tfh = freq[np.argwhere(tpsc_h>=0.5)[0][0]]
tfl = freq[np.argwhere(tpsc_l>=0.5)[0][0]]

plt.figure(figsize=(12,6))      
plt.title('Cumulative sum PS')
plt.ylabel('Cum_sum')
plt.grid(alpha=0.75)
plt.xlabel('freq')    
#	plt.xscale('log')
plt.loglog(freq,tpsc_h,label='Template model- Hanford')
plt.loglog(freq,tpsc_l,label='Template model- Livington')
plt.legend()
plt.show();plt.close()

print('\nFreq. correcponding to half PS weight:')
print('Hanford: (Template,data)',tfh,fh)
print('Livingston: (Template,data)',tfl,fl)
```

These frequencies are approximately half of the sampling rate i.e. Nyquist frequency. This is because LIGO sampling rate is decided such that it is double of their expected signal frequencies. Expected signal frequencies are in the range **(20,2000)Hz** and thus sampling at rate -> 4096Hz allows us to retain all the information about the data. Since the main signal lies in the (20,2000)Hz range we expect similar distribution of power in double of that frequency. Hence because of this, we end up with the half-point frequency close to the **Nyquist frequency (2048)**.

## Part F

Now that we have observed the signal we next move on to using it to locate position of the event. This is done by computing the difference in arrival time of the gravitational wave at the two detectors. This is then translated to a verticle angle which gives a circul in sky where we can look for the event.

For a spacing (L) of couple thousand km,(here assuming 5000km) the verticle angle can be calculated using following equation:
\begin{equation}
\Delta t = L sin(\theta) / c
\end{equation}
Hence,
\begin{equation}
\theta = arcsin(c \Delta t /L)
\end{equation}

We can express the uncertanity in angle in terms of the uncertainity in time. (technically I should to a gaussian fit, but due to a very distinct peak I am taking the error in time to be simply one sample period). Since we are defining $\Delta t$ as difference in time corresponding to the maxima's in the two detector data, net uncertainity on time is 2*dt.
\begin{equation}
d\theta =  (2 dt)L cos(\theta)/ c
\end{equation}


```python
# 	PART F 
# ---------------------------------------------------------------------------
	
# FOr determination of the angular postion range I use the difference in time of
# arrival and translate it to path difference based on the separation of the detectors

dist_detectors = 5000.0*10**3 # meters
speed_wave = 3.0*10**8
delta_t = np.abs(np.argmax(snr_h) - np.argmax(snr_l))*dt_h
# For a vertical angle theta, difference in arrival time time_a = dist_detectors*sin(theta)/speed
# We can get the time difference from the results of the matched fiter 
theta = np.arcsin(delta_t*speed_wave/dist_detectors)
#	delta_t = dist_detectors*cos(theta)/speed *d_theta
d_theta = 2*dt_h*speed_wave/(dist_detectors*np.cos(theta)) # Assuming an uncertainity in time to be 2*dt = 2*0.000244140625s as dt is the least count here for timing data

print('\nDIfference in arrival time (seconds):', delta_t)
print('Uncertainity in verticle angle (degrees):',d_theta*180/np.pi)

# Angular uncertainity is aruond 2degrees. This is not so good in terms positional
# uncertainity as the further we look for the event, worse the uncertanity get.
print('\nIn terms of error in position of event,\nfor distance x, uncertainity is rougly',d_theta,'*x\n\n')	
```

The above results show us that the uncertainity in the vertical angle is appx. 2 degrees. What this means in terms of our position is that instead of a circle we need to look at an annular region of width roughly 3.5% of the distance of ring from us.
