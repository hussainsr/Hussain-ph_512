```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams.update({'font.size': 18})
```

Here, I use convolution with a **delta function** to shift the array being passed as argument to **shift_conv** 

```python
def shift_conv(arr, shift):
	# Takes and array 'arr'  and shifts it by 'shift' by convolution with a delta function carrying 1 at location of shift
	k 		  = np.asarray([0 for i in range(len(arr))])
	k[shift]  = 1. # defining the delta function
	phase_k   = np.fft.fft(k)
	phase_arr = np.fft.fft(arr)
	return np.fft.ifft(phase_arr*phase_k) # returning inv.fourier of the product of the two function's fourier transform
```

Now, defining the array such that the gaussian is centered with the array and then shifting it by half (here, 100/2) the array length 

```python
x 		= np.linspace(-1,1,100)
y 		= np.exp(-x**2) 		  # Unit mean, variance gaussian
y_shift = shift_conv(y,len(y)//2) # Shifting the gaussian by quarter of array size
print('\n(Length of array, shift): ',[100,100/2])
```

```python
plt.figure(figsize=(12,6))      
plt.title('Array shift')
plt.xlabel('x')
plt.grid(alpha=0.75)
plt.ylabel('f(x)')
plt.plot(x,y,label='original gaussian')
plt.plot(x,np.abs(y_shift),label='shifted gaussian')
plt.legend()
plt.show()
plt.close()
```
