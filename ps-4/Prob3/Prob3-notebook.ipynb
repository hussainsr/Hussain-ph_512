```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 18})
```

Using the **shift_conv** function from before

```python
def correlation(arr1,arr2):
	return np.fft.ifft(np.fft.fft(arr1)*np.conj(np.fft.fft(arr2)))

def shift_conv(arr, shift):
	# Takes and array 'arr'  and shifts it by 'shift' by cpmvolution with a delta function carrying 1 at location of shift
	k 		  = np.asarray([0 for i in range(len(arr))])
	k[shift]  = 1 # defining the delta function
	phase_k   = np.fft.fft(k)
	phase_arr = np.fft.fft(arr)
	return np.fft.ifft(phase_arr*phase_k) # returning inv.fourier of the product of the two function's fourier transform
```

Defining the array like before, and using a random number to shift

```python
# Defining the array
x = np.linspace(-1,1,100)
y = np.exp(-x**2) # Gaussian (centered at x=0 with unit variance)
shift_rand = int(np.random.rand()*100)
y_shift = shift_conv(y,shift_rand) # Shifting the gaussian by random number
print('\n(Array size, shift)', [len(x),shift_rand])

plt.figure(figsize=(12,6))      
plt.title('Array shift')
plt.xlabel('x')
plt.grid(alpha=0.75)
plt.ylabel('f(x)')
plt.plot(x,y,label='original gaussian')
plt.plot(x,np.abs(y_shift),label='shifted gaussian')
plt.legend()
plt.show()
plt.close()
```

```python
corr = correlation(y_shift,y_shift)

plt.figure(figsize=(12,6))      
plt.title('Correlation of shifted gaussian with itself (using fft)')
plt.xlabel('x')
plt.grid(alpha=0.75)
plt.ylabel('f(x)')
plt.plot(corr,label = 'Correlation result')
plt.legend()
plt.show()
plt.close()
```

Nothing surprising here, we get back the same plot from before. The correlation is independant of the shift since we are taking the correlation of the shifted gaussain with itself. We will start noticing changes if we were to take the correlation between shifted and normal gaussian.

```python
corr2 = correlation(y,y_shift)

plt.figure(figsize=(12,6))      
plt.title('Correlation of gaussian with shifted gaussian (using fft)')
plt.xlabel('x')
plt.grid(alpha=0.75)
plt.ylabel('f(x)')
plt.plot(corr2,label = 'Correlation result')
plt.legend()
plt.show()
plt.close()
```

This now shows some dependance on the shift because of the presence of relative shift between the two function which was absent before.
